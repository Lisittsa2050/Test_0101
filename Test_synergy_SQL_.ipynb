# Задача № 1

- Выведите только те строки (интересуют все поля таблицы + те, которые указаны ниже), где почта оканчивается на @synergy.ru
- Добавьте столбец, который пронумерует оплаты по их дате в разрезе каждого клиента (для каждого нового клиента его первая оплата начинается с 1)
- Добавьте столбец, который будет отображать месяц совершения оплаты (если оплата была 01.01.2022, то пишем "Январь")
- Отсортируйте итоговые данные по лиду и номеру его оплаты

```python
# создаем таблицу
contactsclients = pd.DataFrame({
                                'id_lead': [1, 1, 1, 2, 2, 3, 3, 4, 4, 4],
                                'mail': ['max@synergy.ru', 'max@synergy.ru', 'max@synergy.ru',
                                         'dron@mail.ru', 'dron@mail.ru', 'dron@synergy.ru',
                                         'dron@synergy.ru', 'max@mail.ru', 'max@mail.ru', 'max@mail.ru'],
                                'date_pay': ['2022-01-03', '2022-01-05', '2022-01-04',
                                             '2022-02-10', '2022-02-11', '2022-03-21',
                                             '2022-04-20', '2022-03-04', '2022-03-10',
                                             '2022-03-10']})

contactsclients
```

```python
def SELECT(sql):
  return pd.read_sql(sql,con)
```

```python
# проверка
sql = '''
SELECT *
FROM contactsclients
'''
SELECT(sql)
```

```python
# пропишем все 12 мес.
sql = '''
WITH Jan AS (
SELECT
      *,
      ROW_NUMBER() OVER (PARTITION BY id_lead ORDER BY date_pay) AS 'номер_оплаты',
      CASE
           WHEN strftime('%m', date_pay) = '01' THEN 'Январь'
           WHEN strftime('%m', date_pay) = '02' THEN 'Февраль'
           WHEN strftime('%m', date_pay) = '03' THEN 'Март'
           WHEN strftime('%m', date_pay) = '04' THEN 'Апрель'
           WHEN strftime('%m', date_pay) = '05' THEN 'Май'
           WHEN strftime('%m', date_pay) = '06' THEN 'Июнь'
           WHEN strftime('%m', date_pay) = '07' THEN 'Июль'
           WHEN strftime('%m', date_pay) = '08' THEN 'Август'
           WHEN strftime('%m', date_pay) = '09' THEN 'Сентябрь'
           WHEN strftime('%m', date_pay) = '10' THEN 'Октябрь'
           WHEN strftime('%m', date_pay) = '11' THEN 'Ноябрь'
           WHEN strftime('%m', date_pay) = '12' THEN 'Декабрь'
           ELSE 'Месяц неопределен'
      END AS 'месяц_оплаты'
FROM contactsclients)

SELECT *
FROM Jan
WHERE mail LIKE '%@synergy.ru' AND месяц_оплаты = 'Январь' -- посмотрим месяц_оплаты только Январь
ORDER BY id_lead, номер_оплаты
'''
SELECT(sql)
```

```python
# все месяца, где почта оканчивается на @synergy.ru
sql = '''
WITH vse AS (
SELECT
      *,
      ROW_NUMBER() OVER (PARTITION BY id_lead ORDER BY date_pay) AS 'номер_оплаты',
      CASE
           WHEN strftime('%m', date_pay) = '01' THEN 'Январь'
           WHEN strftime('%m', date_pay) = '02' THEN 'Февраль'
           WHEN strftime('%m', date_pay) = '03' THEN 'Март'
           WHEN strftime('%m', date_pay) = '04' THEN 'Апрель'
           WHEN strftime('%m', date_pay) = '05' THEN 'Май'
           WHEN strftime('%m', date_pay) = '06' THEN 'Июнь'
           WHEN strftime('%m', date_pay) = '07' THEN 'Июль'
           WHEN strftime('%m', date_pay) = '08' THEN 'Август'
           WHEN strftime('%m', date_pay) = '09' THEN 'Сентябрь'
           WHEN strftime('%m', date_pay) = '10' THEN 'Октябрь'
           WHEN strftime('%m', date_pay) = '11' THEN 'Ноябрь'
           WHEN strftime('%m', date_pay) = '12' THEN 'Декабрь'
           ELSE 'Месяц неопределен'
      END AS 'месяц_оплаты'
FROM contactsclients)

SELECT *
FROM vse
WHERE mail LIKE '%@synergy.ru'
ORDER BY id_lead, номер_оплаты
'''
SELECT(sql)
```

# Задача № 2

- Таблица "Leads" содержит id лидов (id_lead) их дату (date_lead) и поле, связывающее лид с его источником (id_source_lv1).
- Таблица "Invoices" содержит id счёта (id_invoice), id лида (id_lead) и сумму оплат (sum).
- Таблица "source" содержит id источника верхнеуровневое (id_source_lv1), id источника нижнеуровневое (id_source_lv2), название источника верхнеуровневое (source_name_lv1) и название источника нижнеуровневое (source_name_lv2).
- Нужно вывести дату лида, id лида, сумму всех оплат лида, название источника (только lv1).


**Cоздаем таблицы**

```python
leads = pd.DataFrame({
                       'id_lead': ['AAA', 'BBB'],
                       'date_lead': ['2021-01-01', '2022-01-02'],
                       'id_source': ['123', '222']})
leads
```

```python
invoices = pd.DataFrame({
                       'id_invoice': ['inv1', 'inv2', 'inv3', 'inv4'],
                       'id_lead': ['AAA', 'AAA', 'BBB', 'BBB'],
                       'sum_pay': ['100', '50', '30', '30' ]})
invoices
```

```python
source = pd.DataFrame({
                       'id_source_lv1': ['123', '123', '222'],
                       'source_name_lv1': ['Веб', 'Веб', 'Вхдящий звонок'],
                       'id_source_lv2': ['2232', '2442', 'Null'],
                       'source_name_lv2': ['Сайт 1', 'Сайт 2', 'Null']})

source
```

```python
leads.to_sql('leads', con, index = False, if_exists = 'replace')
invoices.to_sql('invoices', con, index = False, if_exists = 'replace')
source.to_sql('source', con, index = False, if_exists = 'replace')
```

```python
sql = '''
SELECT
      leads.date_lead,
      leads.id_lead,
      SUM(invoices.sum_pay) AS total_paid,
      (
        SELECT source_name_lv1
        FROM source
        WHERE id_source_lv1 = leads.id_source
        LIMIT 1
      ) AS source_name_lv1
FROM leads
LEFT JOIN invoices ON leads.id_lead = invoices.id_lead
GROUP BY leads.id_lead, leads.date_lead
'''
SELECT(sql)
```

# Задача № 3

- Есть таблица по перемещению лида между департаментами. У лида есть id, порядковый номер обрабатывающего департамента и название обрабатывающего.
- Необходимо для каждого лида вывести путь перемещения лида начиная с первого департамента.

# Задача № 4

- Есть таблица с суммой оплат (sum_pay) по лидам на каждую дату (date_pay).
- Необходимо вывести сумму оплат накопительным итогом на каждую дату оплаты для каждого лида.

```python
# создаем таблицу
leadpaymentamount = pd.DataFrame({
                            'id_lead': [1, 1, 1, 2, 2, 3, 4, 4, 4],
                            'date_lead': ['2022-01-01', '2022-01-01', '2022-01-01',
                                          '2022-02-01', '2022-02-01', '2022-03-01',
                                          '2022-03-04', '2022-03-04', '2022-03-04'],
                            'date_pay': ['2022-01-03', '2022-01-05', '2022-01-04',
                                         '2022-02-10', '2022-02-11', '2022-03-21',
                                         '2022-03-04', '2022-03-10', '2022-03-10'],
                            'sum_pay': [100, 200, 50, 120, 140, 10, 30, 40, 50]})

leadpaymentamount
```

```python
leadpaymentamount.to_sql('leadpaymentamount', con, index = False, if_exists = 'replace')
```

```python
# решать будем через оконки
sql = '''
SELECT
      id_lead,
      date_pay,
      SUM(sum_pay) OVER (PARTITION BY id_lead	ORDER BY date_pay rows
                         BETWEEN unbounded preceding AND current row) AS Сумма_накопительным_итогом
FROM leadpaymentamount
ORDER BY 1
'''
SELECT(sql)
```
